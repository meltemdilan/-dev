```python
# Veri işleme kütüphaneleri
import pandas as pd  # Veri yükleme, işleme ve manipülasyonu için
import numpy as np  # Sayısal işlemler ve veri analizi için

# Görselleştirme kütüphaneleri
import matplotlib.pyplot as plt  # Grafik oluşturma için temel kütüphane
import seaborn as sns  # Estetik grafikler için matplotlib'in üstüne kurulu
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
# Veri setini yükle
df = pd.read_csv("synthetic_healthcare/medical_data.csv")

# Veri setinin boyutlarını öğren (kaç satır ve sütun var)
print("Veri seti boyutları:", df.shape)

# İlk birkaç satırı göster (ilk 5 satırı görmek için)
print("İlk 5 satır:")
print(df.head())

# Sütun isimlerini listele
print("\nSütun isimleri:", df.columns.tolist())

# Verinin genel bilgilerini göster (sütun adları, veri tipleri, eksik değerler)
print("\nVeri setinin genel bilgileri:")
print(df.info())
```

```python
# Tüm sayısal sütunların özet istatistikleri
print("\nSayısal sütunların özet istatistikleri:")
print(df.describe())

# Eksik değerleri kontrol et
print("\nHer sütunda eksik veri sayısı:")
print(df.isnull().sum())

# Benzersiz değer sayısını kontrol et
print("\nHer sütundaki benzersiz değer sayısı:")
print(df.nunique())
```

```python
# Sadece sayısal sütunları seç
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Histogram çiz
numeric_columns.hist(bins=20, figsize=(15, 10))
plt.suptitle("Histogramlar")
plt.show()
```
